```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def count_images(base_dir):
    counts = {}
    for cls in os.listdir(base_dir):
        cls_path = os.path.join(base_dir, cls)
        if os.path.isdir(cls_path):
            counts[cls] = len(os.listdir(cls_path))
    return counts

train_counts = count_images("dataset/train/")
val_counts = count_images("dataset/val/")
test_counts = count_images("dataset/test/")

print("Train:", train_counts)
print("Val:", val_counts)
print("Test:", test_counts)
```

```python
plt.bar(train_counts.keys(), train_counts.values())
plt.title("Training Dataset Class Distribution")
plt.ylabel("Number of Images")
plt.show()
```

```python
def show_samples(base_dir, cls, n=5):
    cls_dir = os.path.join(base_dir, cls)
    images = os.listdir(cls_dir)[:n]

    plt.figure(figsize=(10, 3))
    for i, img_name in enumerate(images):
        img_path = os.path.join(cls_dir, img_name)
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        plt.subplot(1, n, i+1)
        plt.imshow(img, cmap="gray")
        plt.axis("off")
    plt.suptitle(f"Sample images â€“ {cls}")
    plt.show()

show_samples("dataset/train", "defected")
show_samples("dataset/train", "undefected")
```
